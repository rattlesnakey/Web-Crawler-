```python
#利用session 对象来维护一个会话，不需要先post登录再get登陆后的个人信息两次操作这么麻烦，且这个第二次还是要复制第一次post的cookie
#因为每一次请求你都相当于重新开一个浏览器去访问也就是重新开了一次会话
#用session可以解决这样的问题，比如下面这样
```

```python
requests.get("http://httpbin.org/cookies/set/number/123456789")
r=requests.get("http://httpbin.org/cookies")
print(r.text)#第二次请求的时候重新打开了一个一个浏览器了，重新开了一个会话，就无法知道我刚刚设置的那个cookes：123456789，就会返回空
```

```python
s=requests.Session()#利用session可以模拟同一个对话而不用担心cookie的问题，通常用于模拟登录成功之后的操作
s.get("http://httpbin.org/cookies/set/number/123456789")
r=s.get("http://httpbin.org/cookies")#用一个会话对象来装所有请求的信息，包括请求所返回回来的信息
print(r.text)
```
